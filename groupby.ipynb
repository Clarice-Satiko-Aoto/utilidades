## Flights
>informações gerais sobre os voos


1.   year, month, day: Date of departure
2.   dep_time, arr_time: Actual departure and arrival times (format HHMM or HMM), local tz.
3. sched_dep_time, sched_arr_time: Scheduled departure and arrival times (format HHMM or HMM), local tz.
4. dep_delay, arr_delay: Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.
5. carrier: Two letter carrier abbreviation. See airlines to get name.
6. flight: Flight number.
7. tailnum: Plane tail number. See planes for additional metadata.
8. origin, dest: Origin and destination. See airports for additional metadata.
9. air_time: Amount of time spent in the air, in minutes.
10. distance: Distance between airports, in miles.
11. hour, minute: Time of scheduled departure broken into hour and minutes.
12. time_hour: Scheduled date and hour of the flight as a POSIXctdate. Along with origin, can be used to join flights data to weather data.

### which airline is the most punctual?
> dep_delay, arr_delay: Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.

```python
#Calculando o total de atraso
df_flights['total_delay'] = df_flights['arr_delay'] + df_flights['dep_delay']
#O pontual significa os menos atrasados ou contabiliza aqueles que não tem o atraso?
punctual_airlines = (df_flights
                     .groupby('carrier')
                     .agg(mean_delay=('total_delay','mean'))
                     .sort_values(by='mean_delay')
                     .reset_index()
                     )
punctual_airlines.head()
```

```python
#Merge para descobrir o nome da companhia aerea (df_airlines)
punctual_airlines.merge(df_airlines,how='left')
```

### what destination is the longest duration

```python
#sabe-se que o voo mais longo tem a duração de 695 minutos
df_flights.air_time.describe()
```

```python
# Agrupar os dados para verificar o tempo médio dos voos
longest_duration = (df_flights
                    .groupby(['origin','dest'])
                    .agg(average_air_time=('air_time','mean'))
                    .reset_index()
                    .sort_values(by='average_air_time', ascending=False)
                    )
longest_duration.head(2)
```

```python
# merge com df_airport para saber o nome do aeroporto de destino.
(longest_duration
        .merge(df_airports[['faa','name']], 
               how='left', 
               left_on='dest', 
               right_on='faa')
        )
```

### which airline has the highest capacity of seats?

```python
carrier_tail_num = df_flights[['carrier','tailnum']].drop_duplicates()
carrier_tail_num.head(2)
```

```python
#merge do carrier_tail_num e df_planes para obter o número de assentos,
#agrupar por carrier e somar os assentos,
#classificação decrescente da soma dos assentos
seats = (carrier_tail_num
         .merge(df_planes[['tailnum','seats']], how='left',)
         .groupby(by='carrier')
         .agg(total_seats=('seats','sum'))
         .sort_values(by='total_seats', ascending=False)
         )
seats.head()
```

### which airplane model is the highest in use and from which manufacturer?

```python
#agrupando para contar os valores de tailnum de df_flights
airplane_use = (df_flights
                .groupby('tailnum')
                .agg(count=('tailnum','count'))
                .reset_index())
airplane_use.head(2)
```

```python
#merge com df_planes para conhecer o total de voos por modelo e fabricante
highest_model = (airplane_use
                 .merge(df_planes[['tailnum','model','manufacturer']])
                 .groupby(['model','manufacturer'])
                 .agg(total_flights=('count','sum'))
                 .sort_values(by='total_flights', ascending=False))
highest_model.head(3)
```
